### Análisis de precios de propiedades

Para ver el comportamiento de los precios de las propiedades es una buena opción hacer uso de un histograma, ya que:
- Podemos ver cómo se distribuyen los precios de las propiedades en el mercado.
- Identificamos los rangos de precios más comunes.
- Detectamos si hay propiedades con valores atípicamente altos o bajos.
- Determinamos si la distribución es simétrica o presenta algún sesgo. 

````python
plt.figure(figsize=(10, 6))
sns.histplot(df['Precio'], bins=30, kde=True)
plt.title('Distribución de Precios de Propiedades')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.axvline(df['Precio'].mean(), color='red', linestyle='--', label=f'Media: {df["Precio"].mean():,.0f}')
plt.axvline(df['Precio'].median(), color='green', linestyle='--', label=f'Mediana: {df["Precio"].median():,.0f}')
plt.legend()
plt.show()
````

````python
# Establecer el tamaño del gráfico
plt.figure(figsize=(12, 6))

# Boxplot para visualizar el comportamiento del precio
sns.boxplot(x=df['Precio'])

# Títulos y etiquetas
plt.title('Boxplot de Precios de Propiedades', fontsize=16)
plt.xlabel('Precio (en millones)', fontsize=14)

# Mostrar el gráfico
plt.show()
````

El percentil 99 es el valor por debajo del cual se encuentra el 99% de los datos.

Esto implica que el 1% restante representa los valores más altos (posiblemente valores atípicos o extremos).

````python
plt.figure(figsize=(10, 6))
# Filtramos el 99% de los datos para eliminar extremos
precio_99 = np.percentile(df['Precio'], 99)
sns.histplot(df[df['Precio'] <= precio_99]['Precio']/1000000, bins=30, kde=True)
plt.title('Distribución de Precios de Propiedades (sin valores extremos)')
plt.xlabel('Precio (millones)')
plt.ylabel('Frecuencia')
plt.show()
````

Vemos que la gráfica está sesgada sobre la derecha, esto nos indica que hay más casas que están valiendo menos de un millón o mil millones

````python
# Crear categorías de precios en millones
bins = [0, 200000000, 400000000, 600000000, 800000000, 1000000000000, df['Precio'].max()]
labels = ['0-200M', '200-400M', '400-600M', '600-800M', '800M-1B', '>1B']

# Crear una nueva columna con las categorías
df['Categoría_Precio'] = pd.cut(df['Precio'], bins=bins, labels=labels, right=False)

# Graficar la distribución de propiedades por categoría de precio
plt.figure(figsize=(12, 6))
sns.countplot(x='Categoría_Precio', data=df, palette='viridis')
plt.title('Distribución de Propiedades por Rango de Precio', fontsize=16)
plt.xlabel('Rango de Precio', fontsize=14)
plt.ylabel('Número de Propiedades', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Mostrar el porcentaje de propiedades en cada categoría
porcentajes = df['Categoría_Precio'].value_counts(normalize=True).sort_index() * 100
print("Distribución porcentual de propiedades por categoría de precio:")
for categoria, porcentaje in porcentajes.items():
    print(f"{categoria}: {porcentaje:.1f}%")
````

## Relación entre área y precio con base al tipo de propiedad

````python
# Eliminar valores atípicos tanto en Precio como en Área usando el método IQR

# Calcular percentiles para Precio
precio_99 = np.percentile(df['Precio'], 99)

# Calcular percentiles para Área
area_99 = np.percentile(df['Area'].dropna(), 99)

# Filtrar los datos eliminando valores extremos en ambas variables
df_filtrado = df[(df['Precio'] <= precio_99) & (df['Area'] <= area_99)]

# Crear un gráfico de dispersión mejorado
plt.figure(figsize=(12, 8))

# Gráfico de dispersión con colores según tipo de propiedad 
n_types = len(df_filtrado['Tipo_Propiedad'].unique())
palette = sns.color_palette("Spectral", n_types)

scatter = sns.scatterplot(
    data=df_filtrado,
    x='Area',
    y='Precio',
    hue='Tipo_Propiedad',
    alpha=0.7,
    palette=palette
)

# Convertir valores del eje y a millones
current_values = scatter.get_yticks()
scatter.set_yticklabels(['{:,.0f}'.format(x/1000000) for x in current_values])

# Títulos y etiquetas
plt.title('Relación entre Área y Precio (sin valores atípicos)', fontsize=18)
plt.xlabel('Área (m²)', fontsize=14)
plt.ylabel('Precio (millones)', fontsize=14)

# Ajustar leyenda
plt.legend(title='Tipo de Propiedad', bbox_to_anchor=(1.05, 1), loc='upper left')

# Ajustar diseño
plt.tight_layout() # para evitar que se corten etiquetas
plt.grid(True, alpha=0.3)
plt.show()

# Verificar si df_filtrado no está vacío antes de calcular min/max
if not df_filtrado.empty:
    print(f"Rango de precio: {df_filtrado['Precio'].min()/1000000:.2f} a {df_filtrado['Precio'].max()/1000000:.2f} millones")
else:
    print("No hay datos en el rango filtrado.")
print(f"Datos originales: {len(df)} propiedades")
print(f"Datos filtrados: {len(df_filtrado)} propiedades ({len(df_filtrado)/len(df)*100:.1f}% del total)")
print(f"Rango de área: {df_filtrado['Area'].min():.0f} a {df_filtrado['Area'].max():.0f} m²")
print(f"Rango de precio: {df_filtrado['Precio'].min()/1000000:.2f} a {df_filtrado['Precio'].max()/1000000:.2f} millones")
````

Cuando queremos ver la relación entre varias variables a la vez podemos usar pairplot

````python
# Crear una copia del dataframe y añadir una columna para precio en millones
df_plot = df_filtrado.copy()
df_plot['Precio_millones'] = df_plot['Precio'] / 1000000

# Crear el pairplot con la nueva variable
g = sns.pairplot(df_plot,
             vars=['Area', 'Precio_millones', 'Habitaciones', 'Baños'],
             markers='o')

plt.suptitle('Relación entre Variables', y=1.02, fontsize=16)
plt.tight_layout()
plt.show()
````

**¿Qué representa el histograma en la diagonal?**

- La frecuencia de ocurrencia de los valores para una sola variable.

- Cuántas observaciones caen dentro de cada "bin" o intervalo de valor.

Hay una correlación moderada positiva entre:

- Área y Precio: A mayor área, tiende a haber un mayor precio, aunque con mucha dispersión.

- Habitaciones y Baños: Propiedades con más habitaciones suelen tener también más baños.

Correlación débil entre algunas variables:

- Habitaciones y precio no muestran una relación clara. Hay propiedades muy caras con pocas habitaciones, lo que podría indicar propiedades de lujo o comerciales.

````python
# Definir los tipos de propiedad que quieres mostrar (los más frecuentes o relevantes)
tipos_seleccionados = ['Casa en venta', 'Bodega en arriendo', 'Apartamento en arriendo','Casa en arriendo', 'Apartamento en venta']

# Crear una copia del dataframe filtrado por los tipos seleccionados
df_plot = df_filtrado[df_filtrado['Tipo_Propiedad'].isin(tipos_seleccionados)].copy()
df_plot['Precio_millones'] = df_plot['Precio'] / 1000000

# Crear el pairplot con la nueva variable y hue por tipo de propiedad
g = sns.pairplot(df_plot,
             vars=['Area', 'Precio_millones', 'Habitaciones', 'Baños'],
             hue='Tipo_Propiedad',
             palette='viridis',
             markers='o')

# Move the legend outside to the right
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.suptitle('Relación entre Variables por Tipo de Propiedad', y=1.02, fontsize=16)
plt.tight_layout(rect= [0, 0, 0.85, 1])  # Ajustar el rectángulo para evitar que se corten etiquetas
plt.show()
````

**¿Cómo interpretar las distribuciones de la diagonal?**

- Forma de la curva:

    - Si la curva es alta y estrecha, significa que los valores están muy concentrados en un rango específico.

    - Si es ancha o con varias "jorobas", indica más dispersión o presencia de subgrupos.

- Posición de los picos:

    - Dice cuál es el rango de valores más frecuente para cada tipo de propiedad.

    - Por ejemplo, si la curva de "apartamento en arriendo" está concentrada en áreas pequeñas y precios bajos, se refleja en la gráfica.

- Distribuciones distintas por tipo de propiedad:

    - Las "casas en venta" tienden a tener áreas y precios más altos que los otros tipos.

    - Las "bodegas en arriendo" muestran mucha concentración en precios bajos y áreas grandes, con pocas habitaciones y baños.

- Correlación diferenciada:

    - Las relaciones entre variables como área y precio, o habitaciones y baños, varían bastante entre tipos de propiedad.

    - Por ejemplo, la correlación entre habitaciones y baños parece más fuerte en casas que en bodegas.

Ejemplos:

- ¿En qué ciudad están las propiedades más costosas?

````python
# Calcular estadísticas de precio por ciudad y ordenar por precio promedio
precio_por_ciudad = df_filtrado.query('Ciudad_Final != "desconocida"') \
    .groupby('Ciudad_Final')['Precio'] \
    .agg(['mean', 'median', 'max', 'count']) \
    .sort_values('mean', ascending=False)

# Convertir precios a millones para mejor legibilidad
precio_por_ciudad['mean_millones'] = precio_por_ciudad['mean'] / 1000000
precio_por_ciudad['median_millones'] = precio_por_ciudad['median'] / 1000000
precio_por_ciudad['max_millones'] = precio_por_ciudad['max'] / 1000000

# Mostrar resultados con formato
print("Estadísticas de precios por ciudad:")
display(precio_por_ciudad[['count', 'mean_millones', 'median_millones', 'max_millones']])

# Visualizar los resultados
plt.figure(figsize=(12, 6))
sns.barplot(x=precio_por_ciudad.index, y=precio_por_ciudad['mean_millones'], palette='viridis')
plt.title('Precio Promedio de Propiedades por Ciudad', fontsize=16)
plt.xlabel('Ciudad', fontsize=14)
plt.ylabel('Precio Promedio (millones)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

# Añadir etiquetas con los valores
for i, v in enumerate(precio_por_ciudad['mean_millones']):
    plt.text(i, v + 10, f'{v:.1f}M', ha='center', fontsize=11)

plt.tight_layout()
plt.show()
````
